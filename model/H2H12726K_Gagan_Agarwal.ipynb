## 17. Create a new column in the existing dataframe as “'amount_in_usd'” and convert all the non-USD currencies in USD and store them in the same column.

```python
df["AMOUNT_IN_USD"] = df["ORDER_AMOUNT"].astype(float)
columns = ['CUSTOMER_ORDER_ID','ORDER_AMOUNT','ORDER_CURRENCY','AMOUNT_IN_USD']
precolumn = ['CUSTOMER_ORDER_ID','ORDER_AMOUNT','ORDER_CURRENCY']
print(df[precolumn].sample( n=20))
def convert_to_usd(row): 
    conversion_factors = {
        "USD": 1.0,
        "EUR": 1.07,  
        "AUD": 0.65,  
        "CAD": 0.74,
        "GBP": 1.24,  
        "MYR": 0.21,   
        "PLN": 0.24,   
        "AED": 0.27,     
        "HKD": 0.13,    
        "CHF": 1.11,     
        "RON": 0.22,    
        "SGD": 0.74,     
        "CZK": 0.045,    
        "HU1": 0.0028,    
        "NZD": 0.61,       
        "BHD": 2.65,       
        "SAR": 0.27,       
        "QAR": 0.27,      
        "KWD": 3.25,        
        "SEK": 0.093,  
    }
    curr_currency = row['ORDER_CURRENCY']
    amount = row["AMOUNT_IN_USD"]
    # print("ebhjwgeuguwe",curr_currency)

    if curr_currency == 'USD':
        return amount
    elif curr_currency in conversion_factors:
        factor = conversion_factors[curr_currency]
        converted_currency = amount * factor
        return converted_currency
    else:
        return "NA"

df['AMOUNT_IN_USD'] = df.apply(convert_to_usd,axis=1)
print(df[columns].sample( n=20))

```

```python
currency_counts = df['ORDER_CURRENCY'].value_counts()


threshold = 10000 
top_currencies = currency_counts[currency_counts >= threshold]
other_count = currency_counts[currency_counts < threshold].sum()


top_currencies = top_currencies.append(pd.Series({'Other': other_count}))

colors = plt.cm.Set3(range(len(top_currencies)))

fig, ax = plt.subplots(figsize=(10, 10))

patches, _, _ = ax.pie(top_currencies, labels=top_currencies.index, autopct='%.0f%%', colors=colors, labeldistance=1.1)


for patch in patches:
    label = patch.get_label()
    count = top_currencies[label]
    label_text = f"{label} ({count})"
    patch.set_label(label_text)
    patch.set_color(patch.get_facecolor())


ax.set_title('Currency Distribution')
currency_table = pd.DataFrame({'Currency': currency_counts.index, 'Frequency Count': currency_counts.values})
table = ax.table(cellText=currency_table.values, colLabels=currency_table.columns, loc='bottom')

table.scale(1, 1.5)  
table.auto_set_font_size(False) 
table.set_fontsize(10)  

for i, currency in enumerate(currency_table['Currency']):
    if currency in top_currencies.index:
        color = colors[top_currencies.index.get_loc(currency)]
    else:
        color = colors[-1]  
    table[(i + 1, 0)].set_facecolor(color)

plt.show()
```

# DateTime Manipulation

```python
def difference_in_days(melt, lags, ffday, customer_id_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop = True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff']  = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    return melt


lags = 7 
ffday = 1 
df = difference_in_days(df, lags, ffday, 'CUSTOMER_NUMBER', 'AMOUNT_IN_USD')
```
